```python
class Board:
    def __init__(self, board):
        self.all_nums = set([x for y in board for x in y])
        self.wins = [set(x) for x in board] + [set(z) for z in zip(*board)]
        
    def Play(self, num):
        if num in self.all_nums:
            self.all_nums.discard(num)
            for s in self.wins:
                s.discard(num)
                if not s:
                    return True
                
        return False
```

```python
def day_4_part_1():
    all_boards = [Board(x) for x in day_4_boards]
    for num in day_4_nums:
        for b in all_boards:
            if b.Play(num):
                return sum(b.all_nums) * num

%timeit day_4_part_1()
print(day_4_part_1()) #should be 8136
```
